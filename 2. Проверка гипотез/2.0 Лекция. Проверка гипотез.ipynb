## Гипотеза

Наша задача – проверить гипотезу. Разберёмся, что это такое.


Все эксперименты начинаются с гипотезы. Формулировка гипотезы влияет на все дальнейшие действия, поэтому она должна быть сформулирована корректно.

Хорошо сформулированная гипотеза должна отвечать на три вопроса:
- Что меняем?
- На что повлияет?
- Как измеряем?

Рассмотрим эти пункты на примере с дизайном сайта:

1) `Что меняем?` \
Хотим изменить дизайн сайта. \
2) `На что это повлияет?` \
Изменение дизайна повлияет на поведение пользователей, и в конечном результате увеличит выручку. \
3) `Как измеряем?` \
Для измерения будем использовать метрику - средняя выручка с пользователя. Если метрика значимо увеличится, значит гипотеза верна.

Определим, что такое `средняя выручка с пользователя`, так как эта метрика иногда вызывает разногласия в трактовках. \
Чтобы вычислить среднюю выручку с пользователя, нужно:

1) `Подсчитать суммарную выручку для каждого пользователя за время эксперимента`; \
2) `Вычислить среднее от полученных значений`. \
Продемонстрируем на примере, как вычисляется средняя выручка с пользователя. Допустим, группа состоит из двух покупателей. Первый совершил одну покупку на 1000 рублей, а второй совершил две покупки на 500 и 700 рублей. Выручки с покупателей равны 1000 и 1200 рублей. Средняя выручка с покупателя равна 1100 рублей.

Во время A/B эксперимента мы сравниваем отличия метрики между группами, эти отличия называют `эффектом`.


Эффект может измеряться в абсолютных и относительных величинах. Например, если средняя выручка с пользователя увеличилась с 1000 рублей до 1150 рублей, то абсолютный эффект равен 150 рублей,а относительный - 15%.


Красными точками на графике отмечены значения метрики по каждому пользователю экспериментальной (B) группы, синими - в контрольной (A). Вертикальными линиями соответствующих цветов отмечены средние значения в группах.

Мы обсудили, что такое гипотеза в широком смысле. Для проверки гипотез применяется математическая статистика, в которой используются понятия нулевой и альтернативной гипотез. Мы дадим определение, которое чаще всего подразумевается в терминах АБ тестов.



В эксперименте с дизайном мы проверяем гипотезу о равенстве средних значений выручки с пользователя. \
В данном примере `нулевая гипотеза` – средние значения равны, а `альтернативная гипотеза` – средние значения не равны.

`Гипотеза о равенстве средних` – наиболее широко распространённая нулевая гипотеза в A/B тестировании.

## Проверка гипотез

По данным эксперимента мы можем посчитать значения метрик в `контрольной` и `экспериментальной` группах и оценить значение эффекта. Но это лишь `оценка эффекта`, истинное значение нам неизвестно.

Если провести эксперимент и ничего не менять в экспериментальной группе, то `оценка эффекта`, скорее всего, будет отлична от нуля из-за случайности действий пользователей. А на самом деле эффекта нет, так как никакого воздействия оказано не было.


На картинке выше представлены результаты эксперимента. Каждая точка показывает сумму, на которую пользователь совершил покупки за время эксперимента. Синие точки — пользователи контрольной группы, красные точки — пользователи экспериментальной группы. Вертикальная линия — среднее значение выручки с пользователя в соответствующей цвету группе.

На основе данных можем вывод - эффект скорее всего есть, т.к. видны сильные отличия между группами.

На картинке выше результаты другого эксперимента. Тут можно сказать, что отличия между группами малы, такой результат мог получиться случайно. \
Вероятно, эффекта нет.

Для закрепления возьмем результаты другого эксперимента: \
Известно, что метрика в контрольной группе 100 рублей, а в экспериментальной 200 рублей. \
Для того, чтобы узнать относительный эффект, нужно изменение поделить на базовое значение и умножить на 100% \
`(200 - 100) / 100 * 100% = 100%`

Мы привели примеры, в которых можно оценить на глаз, есть ли значимый результат. Но бывают пограничные случаи, когда непонятно, наблюдаемое отличие случайно или вызвано экспериментом.


На картинке мы видим значительное различие между средними значениями групп. Но вместе с этим мы наблюдаем большой разброс внутри групп. Как же понять, есть эффект или нет? Нам нужен `критерий`, согласно которому мы сможем принимать итоговое решение.


Давайте самостоятельно придумаем простейший статистический критерий. Например, если эффект отличается от нуля больше, чем на 20 рублей, то будем отвергать гипотезу о равенстве средних значений и считать, что эффект есть. Но почему именно 20 рублей, а не 10 или 50? На что влияет выбор порогового значения и как его выбрать, разберёмся в следующем пункте.

## Ошибки I и II рода

При оценке эксперимента мы можем совершить ошибку одного из двух типов: ошибку первого рода или ошибку второго рода.


Ошибки I и II рода ещё называют ложноположительным и ложноотрицательным результатом соответственно.


Обе ошибки могут привести к финансовым потерям:

- При ошибке I рода тратим впустую средства на внедрение изменения, которое не имеет эффекта;
- При ошибке II рода упускаем возможность зарабатывать больше, отказываясь внедрять изменение с эффектом.

В идеальном мире было бы хорошо никогда не ошибаться. В теории можно получать большое кол-во точных результатов, но для того потребуется бесконечно много данных. В реальной жизни мы ограничены в ресурсах и можем собрать только ограниченное кол-во информации. Ошибок не избежать, но можно контролировать их вероятность на определенных уровнях.

Допустим, нулевая гипотеза звучит так — «средняя выручка с пользователя не изменилась»,
а альтернативная — «средняя выручка с пользователя изменилась». Средняя выручка с пользователя на самом деле не изменилась. По результатам эксперимента приняли решение отклонить нулевую гипотезу. В данном случае мы совершили ``ошибку I рода`` - сказали, что эффект есть, хотя его нет.

Возьмем другую ситуацию, нулевая гипотеза звучит так — «средняя выручка с пользователя не изменилась»,
а альтернативная — «средняя выручка с пользователя изменилась». Средняя выручка с пользователя на самом деле изменилась. По результатам эксперимента приняли решение отклонить нулевую гипотезу. Здесь ```ошибок нет```, т.к `эффект есть` и мы `приняли альтернативную гипотезу` = `отклонили нулевую гипотезу`. 

Ранее мы придумали простой статистический критерий и хотели понять, как выбрать пороговое значение эффекта. Давайте посмотрим, как пороговое значение влияет на вероятности ошибок I и II рода. Для этого проведём численный эксперимент:

- Из нормального распределения со средним значением 1000 рублей и стандартным отклонением 100 генерируем три выборки размера 50;
- Две из этих выборок назовем A1 и A2 (ожидаемые значения средних этих выборок одинаковы), а к значениям третьей выборки (назовем её B) искусственно добавим эффект в 50 рублей (на это значение увеличиваем каждый из элементов выборки);
- Подсчитываем средние значения в каждой из групп и проверяем наш критерий: верно ли, что средние в A-группах отличаются меньше, чем на 20 рублей? Верно ли, что среднее значение в B группе отличается от среднего в группе A1 больше, чем на 20 рублей? Сохраняем результаты эксперимента;
- Повторим такие эксперименты 1000 раз, а результаты сохраним в отдельной таблице;
- На основании полученных результатов мы можем посчитать доли ошибок первого и второго рода;
- Ошибка первого рода заключается в том, что разница средних групп А1 и А2 отличались более, чем на 20 рублей. То есть мы заявили, что эффект есть, хотя на самом деле его не было. Ошибка второго рода заключается в том, что разница средних групп А1 и B отличались менее, чем на 20 рублей. То есть мы заявили, что эффекта нет, хотя на самом деле он был.

## Уровень значимости и мощность

Обычно ошибку I рода считают более важной, поэтому в первую очередь фиксируют её вероятность. Для этого используется понятие уровня значимости.


Выбор значения уровня значимости зависит от того, насколько критично совершение ошибки I рода. В бизнесе и социальных науках часто используют значение уровня значимости, равное 0.05.

Выбрав уровень значимости, мы можем подобрать пороговое значение эффекта статистического критерия так, чтобы частота ошибок I рода была равна уровню значимости. Сделав это, можно оценить статистическую мощность критерия.


Статистическая мощность и вероятность ошибки II рода в сумме дают единицу. Чем больше мощность, тем меньше вероятность ошибки II рода и наоборот.

Можно придумать много статистических критериев для проверки гипотезы о равенстве средних. Основные критерии мы рассмотрим в нашем курсе. Из всех критериев нужно выбирать тот, который в конкретной задаче при фиксированном уровне значимости показывает наибольшее значение мощности. Такой критерий называют `наиболее мощным` или `наиболее чувствительным`.

## T-test 

Мы создали свой статистический критерий, который проверяет гипотезу с уровнем значимости 5%. Но наш критерий не универсален, он подходит только для тех данных, которые мы сгенерировали. Если изменить стандартное отклонение, размер выборок или размер эффекта, то вероятности ошибок I и II рода изменятся, а значит, критерий перестанет удовлетворять нашим требованиям.

К счастью, математики решили эту проблему и придумали более универсальные статистические критерии, которые хорошо работают для большинства экспериментов.

Для проверки гипотезы о равенстве средних есть критерий, который называется `t-test` (например, популярный тест Стьюдента является t-test'ом). t-test реализован во многих статистических программах. Чтобы проверить гипотезу с помощью t-test'а, достаточно передать программе данные контрольной и экспериментальной групп. Результатом работы теста является число (в будущем будем называть его p-value), по значению которого принимается решение, есть статистически значимый эффект или нет при выбранном уровне значимости.

Формально t-test накладывает на данные некоторые ограничения: независимость или нормальность распределения средних. Но фактически мы можем получать с его помощью корректные результаты в гораздо более широком спектре случаев. Как проверить, подходит статистический тест в вашей ситуации или нет, мы расскажем в следующем уроке.

Для примера возьмем такую ситуацию. Нулевая гипотеза звучит так — «средняя выручка с пользователя не изменилась»,
а альтернативная — «средняя выручка с пользователя изменилась». Средняя выручка с пользователя на самом деле изменилась. По результатам эксперимента приняли решение не отклонять нулевую гипотезу. Мы совершили ошибку II рода, решили, что эффекта нет, хотя он есть.

Хотим проверить гипотезу на уровне значимости 0.05. Получили p-value равное 0.24. Отклоняем нулевую гипотезу или нет? \
Не отклоняем, т.к. p-value больше уровня значимости.

## P-value

Кроме t-test'а есть и другие статистические критерии. \
У разных критериев разные формулы и разные распределения. Вот несколько популярных критериев:

- Тест Стьюдента. Проверяет `равенство средних` значений выборок;
- Тест Манна-Уитни. Проверяет, что две выборки из `одного распределения`;
- Тест Дики-Фуллера. Проверяет `наличие единичного корня` в авторегрессионной модели временного ряда первого порядка;
- Тест Колмогорова-Смирнова. Проверяет `принадлежность выборки` к заранее известному `распределению`.

Чтобы сравнить результаты разных критериев нужно помнить распределения и пороговые значения разных тестов — это неудобно. Чтобы легче было сравнивать результаты разных критериев можно использовать p-value.


Если статистикой эксперимента является разница средних, а в экспериментальной группе среднее на 5 рублей больше, то p-value — это вероятность получить отличие между средними не меньше чем на 5 рублей при условии, что эффекта на самом деле нет.

Нужно отметить, что когда мы хотим отлавливать отклонение среднего как в большую, так и в меньшую сторону, то в качестве статистики выбирается модуль разности между средними.


С помощью p-value легко понимать результат любого статистического критерия:

- Если `p-value меньше` уровня значимости α (обычно берется равным 0.05), то отклоняем нулевую гипотезу, то есть заявляем, что `удалось обнаружить значимые отличия`;
- Если `p-value больше` уровня значимости α (обычно берется равным 0.05), то данные нулевой гипотезе не противоречат, `значимые отличия не найдены`.

`Пример 1.` \
Узнаем статистическую значимость отличия средних между значениями в столбцах A и B на уровне значимости 0.05 с помощью t-test. Отклоняем нулевую гипотезу о равенстве средних или нет?

Т-test показал p-value = 0.029. Значит отклоняем нулевую гипотезу, т.к. p-value меньше уровня значимости (0.05)

`Пример 2.` \
Узнаем статистическую значимость отличия средних между значениями в столбцах A и C на уровне значимости 0.05 с помощью t-test. Отклоняем нулевую гипотезу о равенстве средних или нет?

Т-test показал p-value = 0.347. Значит не отклоняем нулевую гипотезу, т.к. p-value больше уровня значимости (0.05)

`Пример 3.` \

На графике видно, что средняя выручка с пользователя в экспериментальной группе больше (группа B). \

## Итоги

`Гипотеза` — это некоторое предположение, которое требует доказательств.

`Нулевая гипотеза` — предположение о том, что эффекта нет, а альтернативная гипотеза — предположение о том, что эффект есть.

`Статистический критерий` — математическое правило, в соответствии с которым отвергается или не отвергается нулевая гипотеза. Для проверки гипотезы о равенстве средних подходит t-test.

При `проверке гипотез` можно совершить одну из двух ошибок:

- Ошибка I рода — сказать, что эффект есть, когда на самом деле его нет.
- Ошибка II рода — сказать, что эффекта нет, когда на самом деле он есть.

`Уровень значимости` — вероятность того, что статистический критерий совершит ошибку I рода. Часто используют alpha = 0.05.

`p-value` — вероятность получить значение статистики как в эксперименте или более критическое при верности нулевой гипотезы. Если p-value меньше уровня значимости, то нулевая гипотеза отвергается в пользу альтернативной, и говорят, что  есть статистически значимый эффект, иначе оснований отвергать нулевую гипотезу нет.
